```python
def count_by_ideal(gen, n):
    J = P.ideal([gen] + [y^(2^n) + y for y in P.gens()]) #relations from generator Q(x0,x1,x2) and F_2^n points
    return (J.vector_space_dimension() - 1) // (2^n-1)
```
